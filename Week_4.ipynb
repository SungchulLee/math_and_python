```python
# 최성훈
# Cycloid

import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 720) * (np.pi/180)
x = theta - np.sin(theta)
y = 1 - np.cos(theta)

plt.plot(x, y)
plt.xlim([-1,15])
plt.ylim([0,10])
plt.show()
```

```python
# 최성훈
#중력장에서 비스듬하게 던져진 물체의 운동 궤적

import matplotlib.pyplot as plt
import numpy as np

theta = eval(input("발사각 입력 (in degree) ->")) * (np.pi/180)
v = eval(input("발사속력 입력 (in m/sec) ->"))
g = 9.8

c = 2*v*np.sin(theta)/g
t = np.linspace(0, c)

x = np.cos(theta)*v*t
y = np.sin(theta)*v*t -1/2*g*t**2

x_max = np.sin(2*theta)*v**2/g
y_max = np.sin(theta)**2*v**2/(2*g)

plt.plot(x,y)
plt.xlim([0,x_max])
plt.ylim([0,y_max])
plt.show()
```

```python
# 강영수
# sinx, cosx 그래프

import numpy as np
import matplotlib. pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.set_xlim([-3.5, 3.5])
ax.set_ylim([-1.2, 1.2])
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels(['$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$'])

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

x = np.linspace(-np.pi, np.pi)
sin_x = np.sin(x)
cos_x = np.cos(x)

ax.plot(x, sin_x, color='blue', linewidth=2)
ax.plot(x, cos_x, color='red', linewidth=2)

plt.show()
```

```python
# 어정규

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(9)

x = np.arange(10)
y = [x, x**2, np.random.choice(100, size=10)]

styles = ['-', '--', ':']
colors = ['deepskyblue', 'mediumseagreen', 'hotpink']
widths = [10, 5, 2]

plt.figure(figsize=(3,4))

for i in range(len(y)) : 
    plt.plot(x, y[i], \
             linestyle=styles[i], \
             color=colors[i], \
             linewidth=widths[i])  
plt.title("Line examples")
plt.xlabel("X")
plt.ylabel("Y")
plt.ylim([0,100])
    
plt.show()
```

```python
# 어정규

from matplotlib import pyplot as plt
import numpy as np

x = np.arange(10)
y = x*5
y2 = np.tan(x)

plt.plot(x, y, 'vr', label='first')
plt.plot(x, y2, label='second')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('sample')
plt.legend(loc='upper right')

plt.show()
```

```python
# 박소현

import matplotlib.pyplot as plt

days = [0,1,2,3,4,5,6]
max_t = [9,13,13,15,16,17,14]
min_t = [0,1,0,3,6,3,7]
avg_t = [5,7,6,9,11,10,12]

plt.figure()

if 0:
    plt.plot(avg_t)
elif 0:
    plt.plot(days, avg_t)
elif 0:
    plt.plot(days, avg_t, '-oy')
elif 0:
    plt.plot(days, max_t)
    plt.plot(days, min_t)
    plt.plot(days, avg_t)
elif 0:
    plt.plot(days, max_t, label="max")
    plt.plot(days, min_t, label="min")
    plt.plot(days, avg_t, label="average")
    plt.legend()
elif 0:
    plt.plot(days, max_t, '--r', label="max")
    plt.plot(days, min_t, '--k', label="min")
    plt.plot(days, avg_t, '-b', label="average")
    plt.legend()    
elif 1:
    plt.plot(days, max_t, '--r', label="max")
    plt.plot(days, min_t, '--k', label="min")
    plt.plot(days, avg_t, '-b', label="average")
    plt.legend(loc='upper right', shadow=True, fontsize='large')

plt.show()
```
